```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from IPython.display import display

# Enable interactive backend for matplotlib (optional, but recommended for smoother updates)
%matplotlib inline 

# Define the plotting function
def plot_sine_wave(frequency):
    t = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(frequency * t)
    plt.figure(figsize=(8, 4))
    plt.plot(t, y)
    plt.title(f"Sine Wave (Frequency: {frequency:.2f})")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

# Create a FloatSlider widget for frequency control
frequency_slider = FloatSlider(
    value=1.0,
    min=0.1,
    max=5.0,
    step=0.1,
    description='Frequency:',
    continuous_update=True
)

# Use interact to link the slider to the plotting function
interact(plot_sine_wave, frequency=frequency_slider)
```
